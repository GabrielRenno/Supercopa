```python
import pandas as pd
```

```python
atletico = pd.read_html("https://en.wikipedia.org/wiki/2021_Clube_Atl%C3%A9tico_Mineiro_season")
```

```python
atletico[45]
```

```python
time1 = atletico[82][1][0]
placar = atletico[82][2][0]
time2 = atletico[82][3][0]
resultado = time1 + placar  +time2
resultado
```

```python
df= pd.DataFrame()
for i in range(45,83):
    time1 = atletico[i][1][0]
    resultado = atletico[i][2][0]
    time2 = atletico[i][3][0]
    row = pd.Series([time1,resultado,time2])
    row_df = pd.DataFrame([row])
    df = pd.concat([row_df, df], ignore_index=True)


```

```python

#mudar o nome das colunas
df.columns= ["time1","resultado","time2"]
df
```

```python
#separar em duas colunas
new = df["resultado"].str.split("–", n=1, expand=True)
df["Home"] = new[0]
df["Away"] = new[1]
df
```

```python
import numpy as np
```

```python
#Cria colunas de gols pro e gols contra
df['pro'] = np.where(df['time1']== 'Atlético Mineiro', df["Home"], df["Away"])
df['contra'] = np.where(df['time1']!= 'Atlético Mineiro', df["Home"], df["Away"])
```

```python
df.to_csv("atletico.csv", sep='\t', encoding='utf-8') 
```
