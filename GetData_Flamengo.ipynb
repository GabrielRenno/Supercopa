```python
import pandas as pd
```

```python
flamengo = pd.read_html("https://pt.wikipedia.org/wiki/Temporada_do_Clube_de_Regatas_do_Flamengo_de_2021")
```

```python
time1 = flamengo[81][1][0]
placar = flamengo[81][2][0]
time2 = flamengo[81][3][0]
resultado = time1 + placar  +time2
resultado
```

```python
df= pd.DataFrame()
for i in range(81,119):
    time1 = flamengo[i][1][0]
    resultado = flamengo[i][2][0]
    time2 = flamengo[i][3][0]
    row = pd.Series([time1,resultado,time2])
    row_df = pd.DataFrame([row])
    df = pd.concat([row_df, df], ignore_index=True)


```

```python

#mudar o nome das colunas
df.columns= ["time1","resultado","time2"]

```

```python
#separar em duas colunas
new = df["resultado"].str.split(" ", n=1, expand=True)
df["Home"] = new[0]
df["Away"] = new[1]
new = df["Away"].str.split(" ", n=1, expand=True)
df["Away"] = new[1]
df
```

```python
import numpy as np
```

```python
#Cria colunas de gols pro e gols contra
df['pro'] = np.where(df['time1']== 'Flamengo', df["Home"], df["Away"])
df['contra'] = np.where(df['time1']!= 'Flamengo', df["Home"], df["Away"])
```
