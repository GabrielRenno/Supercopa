```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
atletico = pd.read_csv('atletico.csv', sep = "\t")
flamengo = pd.read_csv('flamengo.csv',sep = "\t")
```

```python
atletico.pro.hist()
flamengo.pro.hist()
```

```python
atletico.contra.hist()
flamengo.contra.hist()
```

```python
media_gols_pro_atletico = atletico.pro.mean()
media_gols_pro_flamengo = flamengo.pro.mean()

std_gols_pro_atletico = atletico.pro.std()
std_gols_pro_flamengo = flamengo.pro.std()

media_gols_contra_atletico = atletico.contra.mean()
media_gols_contra_flamengo = flamengo.contra.mean()

std_gols_contra_atletico = atletico.contra.std()
std_gols_contra_flamengo = flamengo.contra.std()


print("Media de gols pró atletico ", media_gols_pro_atletico)
print("Std gols pró atletico ", std_gols_pro_atletico)
print("Media de gols pró flamengo", media_gols_pro_flamengo)
print("Std gols pró flamengo ", std_gols_pro_flamengo)

print("Media de gols contra atletico ", media_gols_contra_atletico)
print("Std gols contra atletico", std_gols_contra_atletico)
print("Media de gols contra flamengo", media_gols_contra_flamengo)
print("Std gols contra flamengo ", std_gols_contra_flamengo)
```

```python
def gameSim():
    atleticoscore = (rnd.gauss(media_gols_pro_atletico,std_gols_pro_atletico)
    +rnd.gauss(media_gols_contra_flamengo,std_gols_contra_flamengo))/2

    flamengoscore = (rnd.gauss(media_gols_pro_flamengo,std_gols_pro_flamengo)
    + rnd.gauss(media_gols_contra_atletico,std_gols_contra_atletico))/2
    
    if int(round(atleticoscore)) > int(round(flamengoscore)):
        return 1
    elif int(round(atleticoscore)) < int(round(flamengoscore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    tie = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
        else: tie +=1 
    print('Atletico Win ', team1win/(team1win+team2win+tie),'%')
    print('Flamengo Win ', team2win/(team1win+team2win+tie),'%')
    print('Tie ', tie/(team1win+team2win+tie), '%')
    return gamesout
```

```python
gamesSim(100000)
```
